```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"D:\computer_version\image\lake.jpg")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("original")
plt.show()
cv2.waitKey(0)
```

```python
image
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
binary=gray.copy()
for i in range(binary.shape[0]):
    for j in range(binary.shape[1]):
        if binary[i][j]>150:
            binary[i][j]=0
        else:
            binary[i][j]=255
plt.imshow(cv2.cvtColor(binary,cv2.COLOR_BGR2RGB))
plt.title("dst_binary")
plt.show()
cv2.waitKey(0)
```

```python
ret, dst_binary = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(dst_binary,cv2.COLOR_BGR2RGB))
plt.title("dst_binary")
plt.show()
cv2.waitKey(0)
```

```python
result = cv2.bitwise_and(image, image, mask=dst_binary)
plt.imshow(cv2.cvtColor(result,cv2.COLOR_BGR2RGB))
plt.title("result1")
plt.show()
cv2.waitKey(0)
```

```python
temp=cv2.subtract(image,result)
plt.imshow(cv2.cvtColor(temp,cv2.COLOR_BGR2RGB))
plt.title("temp")
plt.show()
cv2.waitKey(0)
```

```python
result2=temp.copy()
for i in range(3):
    result2[:,:,i]=cv2.add(dst_binary,result2[:,:,i])
print(result2.shape)
plt.imshow(cv2.cvtColor(result2,cv2.COLOR_BGR2RGB))
plt.title("result2")
plt.show()
cv2.waitKey(0)
```
