```python
from bs4 import BeautifulSoup as bs 
import requests
```

```python
url="https://www.ptt.cc/bbs/Stock/index.html"
```

```python
headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"}
```

```python
res = requests.get(url , headers = headers)
print(res.text)
```

```python
soup = bs(res.text,"lxml")
```

```python
data = soup.select("div.r-ent")
print(data[1])
```

```python
previous_page = soup.select("div#action-bar-container div.btn-group-paging a")[1]["href"]
previous_page = "https://www.ptt.cc" + previous_page
print(previous_page)
```

```python
page_num = previous_page.replace("https://www.ptt.cc/bbs/Stock/index","").replace(".html","")
print(page_num)
```

```python
def get_parsing_data(soup):
    data = soup.select("div.r-ent")
    
    result = []
    
    for sample in data:
        # print(sample)
        
        # 標題
        title = sample.select("div.title")[0].text.strip()

        if "公告" in title or "閒聊" in title or "刪除" in title : 
            continue

        # 時間
        time = sample.select("div.date")[0].text.strip()

        # 推文數量 if>0-->有推文數量
        push_num = sample.select("span.hl")[0].text if len(sample.select("span.hl")) >0 else 0

        # 連結
        raw_link = sample.select("div.title a")[0]["href"]
        domain_name = "https://www.ptt.cc"
        link = domain_name + raw_link
        
        result.append({
            "title" : title , 
            "time"  : time ,
            "push_num" : push_num,
            "link"  : link
        })
        
    return result 
        
get_parsing_data(soup)
```

```python
output = []
for i in range(3):
    url = "https://www.ptt.cc/bbs/Stock/index{}.html".format(int(page_num) - i) 
    res = requests.get(url , headers = headers)
    soup = bs(res.text,"lxml")
    
    result = get_parsing_data(soup)
    output += result 
    
    print("{} is ok.".format(url))

print("Done.")
```

```python
import pandas as pd 
df = pd.DataFrame(output)
df
```

```python
n=len(df.link)
n
```

```python
final=[]
for i in range(n):
    url=df.link[i]
    res = requests.get(url , headers = headers)
    soup = bs(res.text,"lxml")
    header = soup.find_all('span','article-meta-value')
    try:
        title=header[2].text
        time=header[3].text
        title_id=header[0].text
        
        ## 查找所有html 元素 抓出內容
        main_container = soup.find(id='main-container')
        # 把所有文字都抓出來
        all_text = main_container.text
        # 把整個內容切割透過 "-- " 切割成2個陣列
        pre_text = all_text.split('--')[0]
    
        # 把每段文字 根據 '\n' 切開
        texts = pre_text.split('\n')
        # 如果你爬多篇你會發現 
        contents = texts[2:]
        # 內容
        content = '\n'.join(contents)
        content=content.replace('\n', '')
        push_id=soup.find_all("span","f3 hl push-userid")
        push_content=soup.find_all("span","f3 push-content")
        for j in range(len(push_id)):
            all_id=soup.find_all("span","f3 hl push-userid")[j].text
            all_content=soup.find_all("span","f3 push-content")[j].text.replace(": ","")
            all_push_time=soup.find_all("span","push-ipdatetime")[j].text.replace("\n","")
            final.append({
                "title" : title ,
                "title_id":title_id,
                "time"  : time ,
                "content"  : content,
                "push_id":all_id,
                "push_content":all_content,
                "push_time":all_push_time
                })
    except IndexError:
        continue
```

```python
import pandas as pd
df=pd.DataFrame(final)
```
